For clustering

```python
def cluster_and_evaluate(n_clusters, X):

    k_means = KMeans(n_clusters=n_clusters, random_state=1337).fit(X)

    cluster_labels = k_means.predict(X)

    silhouette_avg = silhouette_score(X, cluster_labels)

    print("For n_clusters =", n_clusters)
    print("The average silhouette_score is:", silhouette_avg)
```

### MiniBatchKMeans

```python
# Import of "cluster" for MiniBatchKMeans from sklearn
from sklearn.cluster import MiniBatchKMeans
# Import of numpy
import numpy as np
```

```python
X = buyer_df[["min_buy_price", "max_buy_price", "min_buy_areal", "max_buy_areal", "housing", "hotel", "retail", "office", "land", "industry" ]]
```

```python
def cluster_and_evaluate(n_clusters, X):

    # manually fit on batches
    kmeans = MiniBatchKMeans(n_clusters=5, random_state=0, batch_size=6).fit(X)

    kmeans = kmeans.partial_fit(X)

    kmeans.cluster_centers_

    cluster_labels = kmeans.predict(X)

    silhouette_avg = silhouette_score(X, cluster_labels)

    print("For n_clusters =", n_clusters)
    print("Silhouette average score:", silhouette_avg)

    # fit on the whole data
    kmeans = MiniBatchKMeans(n_clusters=n_clusters, random_state=0, batch_size=6, max_iter=10).fit(X)

    kmeans.cluster_centers_

    cluster_labels = kmeans.predict(X)

    silhouette_avg = silhouette_score(X, cluster_labels)

    print("For n_clusters =", n_clusters)
    print("Silhouette average score:", silhouette_avg)
```

```python
cluster_and_evaluate(5, X)
```
